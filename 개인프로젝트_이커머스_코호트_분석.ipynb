````python
# Íµ¨Îß§ 'Ïõî' ÎΩëÏïÑÎÇ¥Í∏∞
# InvoiceDate Ïª¨ÎüºÏùò ÏµúÏ¥à Íµ¨Îß§ 'Ïõî'ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÏΩîÌò∏Ìä∏Î•º Í∑∏Î£πÌôî ÌïòÍ∏∞ ÏúÑÌï¥ Íµ¨Îß§ 'Ïõî' Ïª¨Îüº Ï∂îÍ∞Ä

# Íµ¨Îß§ÏùºÏãúÎ•º Íµ¨Îß§Ïõî(1Ïùº)Î°ú Î∞îÍøîÏ£ºÎäî Ìï®Ïàò
def get_month(x): return dt.datetime(x.year, x.month, 1)
df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month)

df.head()
````

````python
# Íµ¨Îß§ÏûêÎ≥Ñ ÏµúÏ¥à Íµ¨Îß§ Ïõî Ïª¨Îüº ÌïÑÏöî
# CustomerIDÍ∏∞Ï§Ä, InvoiceMonthÏ§ë Í∞ÄÏû• ÏûëÏùÄ Í∞íÏùÑ CohortMonth Ïª¨ÎüºÏóê Ìï†Îãπ
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceMonth'].transform('min')
df.head()
````

## **Process 02**
#### ÏΩîÌò∏Ìä∏ Î∂ÑÏÑù

#### CohortIndex ÏÉùÏÑ±


---
- pivot table(row: cohort, column: ÏãúÍ∞Ñ Í≤ΩÍ≥º, value: retention rate)
- ÏãúÍ∞Ñ Í≤ΩÍ≥º columnÏùò indexÎ•º CohortIndexÎ°ú ÏßÄÏ†ï

````python
# InvoiceMonth, CohortMonthÏóê Ï†ÅÏö©
# InvoiceMonth - CohortMonth Ïùò Ï∞®Ïù¥Í∞íÏùÄ Í≥ß Ìï¥Îãπ Íµ¨Îß§Í±¥Ïù¥ ÏµúÏ¥à Íµ¨Îß§Î°úÎ∂ÄÌÑ∞ Î™á Í∞úÏõîÏù¥ Í≤ΩÍ≥ºÌñàÎäîÏßÄ ÏïåÏàò ÏûàÎäî ÏßÄÌëúÎ•º ÎÇòÌÉÄÎÇ∏Îã§.
# Ï∞®Ïù¥Í∞íÏùÑ CohortIndexÏóê Ï†ÄÏû•
years_diff = df['InvoiceMonth'].dt.year - df["CohortMonth"].dt.year
months_diff = df['InvoiceMonth'].dt.month - df['CohortMonth'].dt.month
# ÏµúÏ¥à Íµ¨Îß§ÏõîÏùò cohort indexÎ•º 1Î°ú ÎëòÍ≤ÉÏù¥ÎØÄÎ°ú +1 Ï≤òÎ¶¨
df['CohortIndex']= years_diff * 12 + months_diff + 1
df.head()
````

````python
# pivot table ÏÉùÏÑ±

# Í∏∞Ï§Ä Ïõî, Í∑∏ ÏõîÎ°úÎ∂ÄÌÑ∞ ÏßÄÎÇú Í∏∞Í∞ÑÏùò Í≥†Í∞ù ÏàòÎ•º Í≥ÑÏÇ∞
# InvoiceMonth, CohortIndex Í∏∞Ï§ÄÏúºÎ°ú Í≥†Í∞ù Count


cohort_group = df.groupby(['CohortMonth','CohortIndex'])
cohort_df = cohort_group['CustomerID'].nunique()
cohort_df
````

````python
# cohortÎ≥Ñ 1Í∞úÏõîÏî© Í≤ΩÍ≥ºÌï®Ïóê Îî∞Îùº Íµ¨Îß§ÎÇ¥Ïó≠Ïù¥ ÏûàÎäî CustomerIDÏùò ÏàòÎ•º Î≥¥Ïó¨Ï£ºÎäî pivot table
# NullÍ∞íÏùÄ ÏïÑÏßÅ ÎèÑÎûòÌïòÏßÄ ÏïäÏùÄ ÏãúÏ†êÏù¥Ïó¨ÏÑú ÏûêÎ£åÍ∞Ä ÏóÜÎäî Íµ¨Í∞Ñ
cohort_df = cohort_df.reset_index()
cohort_counts = cohort_df.pivot_table(index='CohortMonth',columns='CohortIndex',values='CustomerID')
cohort_counts
````

#### Cohort Î∂ÑÏÑù
- Í≥†Í∞ù Retation(Ïû¨Íµ¨Îß§Ïú®) Î∂ÑÏÑù

````python
# CohortIndexÍ∞Ä 1Ïù∏ Ï≤´Î≤àÏß∏ Ïª¨ÎüºÏù¥ cohortÏùò size
cohort_size = cohort_counts.iloc[:,0] # iloc[:,0] Ï≤´Î≤àÏß∏Ïó¥ÎßåÍ∞ÄÏ†∏Ïò§Í≤†Îã§
# pivot table Î≥∏ÎûòÏùò Í∞íÎì§ÏùÑ cohort sizeÎ°ú ÎÇòÎàÑÎ©¥ ÎπÑÏú®Ïùò ÌòïÌÉúÎ°ú Î≥ÄÌôò
retention = cohort_counts.divide(cohort_size, axis=0) # axis=0 ÌñâÍ∏∞Ï§Ä
retention.round(3)*100 # ÎπÑÏú®(%)Îã®ÏúÑ
````

#### ÏãúÍ∞ÅÌôî (Heatmap)

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

xticks = np.arange(1,14)
yticks = ['2010/12', '2011/01', '2011/02', '2011/03', '2011/04', '2011/05', '2011/06', '2011/07', '2011/08', '2011/09', '2011/10', '2011/11', '2011/12']

plt.figure(figsize=(10,8))
plt.title('Retation rates')
sns.heatmap(data=retention,
            annot = True,
            xticklabels = xticks,
            yticklabels = yticks,
            fmt = '.0%',
            vmin = 0.0,
            vmax = 0.5,
            cmap = 'coolwarm')
plt.show()
````

