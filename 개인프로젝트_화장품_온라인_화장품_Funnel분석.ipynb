## **Process03**  
#### 판매전략 개선


### 상품별 Conversion Rate 분석
---

````python
# 상품별 조회, 장바구니 추가, 구매 데이터 계산
product_events = df.groupby(['product_id', 'event_type']).size().unstack().fillna(0)

# 구매 전환률 계산: 구매수 / 조회수
product_events['conversion_rate'] = product_events['purchase'] / product_events['view'] * 100

# 상품별 조회수가 10회 이상인 상품들만 필터링하여 전환률 TOP 10 확인
top_conversion_products = product_events[product_events['view'] >= 10].sort_values(by=['view', 'conversion_rate'], ascending=[False, True]).head(10)

top_conversion_products[['view', 'cart', 'purchase', 'conversion_rate']]
````

### 일(day)별 행동 패턴분석
---

````python
# event_time에서 일 정보 추출
def get_day(x): return x.day
df['day'] = df['event_time'].apply(get_day)
df.head()
````

````python
# 일별 이벤트 유형별 발생 횟수 계산
daily_events = df.groupby(['day', 'event_type']).size().unstack().fillna(0)

# 일별 구매 전환률 계산: 구매수 / 조회수
daily_events['conversion_rate'] = daily_events['purchase'] / daily_events['view'] * 100

daily_events[['view', 'cart', 'purchase', 'conversion_rate']]
````

````python
# gradient 시각화
styled_all_columns = daily_events.style.background_gradient(cmap='coolwarm', subset=['conversion_rate'])
styled_all_columns
````

### 사용자 행동 패턴분석 (Funnel 분석)
---

````python
# 행동패턴 분석을 위한 Sorting
data_sorted = df.sort_values(['user_session', 'event_time'])
data_sorted.head()
````

````python
# Data가 너무 크므로 가장 일별 구매 전환률(convention rate)이 높은 일(day)을 가져온다
df_day = df[df['day'] == 12]
df_day['event_time'] = pd.to_datetime(df_day['event_time'])
df_day
````

````python
# 최초 view를 시간기준으로 하기위해 min으로정렬
grouped = df_day.groupby(['user_session', 'event_type'])["event_time"].min()
grouped
````

````python
# view, cart, purhase 단계순서를 지정
funnel_steps = pd.DataFrame({'steps':[1,2,3]}, index=['view','cart','purchase'])
funnel_steps
````

````python
# Funnel 단계와 merge
grouped = pd.DataFrame(grouped).merge(funnel_steps, left_on = 'event_type', right_index = True)
grouped
````

````python
# pivot table를 이용하여 user session별 Funnel 단계 진행여부를 확인
funnel = grouped.reset_index().pivot(index='user_session',columns='steps',values='event_time')
funnel.columns = funnel_steps.index
funnel
````

````python
step_values = [funnel[column].notnull().sum() for column in funnel.columns]
step_values
````

### Funnel 분석 시각화
---

````python
import plotly.express as px
data = dict(
    number = [28987, 6289, 1192],
    stage = ["view","cart","purchase"])
fig = px.funnel(data, x='number',y='stage')
fig.show()
````

view->cart 단계에서 사용자 수가 많이 감소하지만,

▶ Data가 너무 커서 전환률이 가장 높은 일자를 기준으로 Funnel 분석을 진행하였는데,
만약 전체 Data를 기준으로 진행하였다면 결과가 어떤식으로 다를지 궁금하다.
