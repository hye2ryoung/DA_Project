````python
# Ïù¥Î≤§Ìä∏ ÌÉÄÏûÖ ÌöüÏàò
event_counts = df['event_type'].value_counts()
event_counts
````

````python
# Í∞Å Îã®Í≥ÑÏóêÏÑúÏùò Ïù¥ÌÉàÎ•†
view_to_cart_dropout = (event_counts['view'] - event_counts['cart']) / event_counts['view'] * 100
cart_to_purchase_dropout = (event_counts['cart'] - event_counts['purchase']) / event_counts['cart'] * 100
cart_to_remove_dropout = (event_counts['remove_from_cart']) / event_counts['cart'] * 100

dropout_rates = {
    'View to Cart Dropout Rate (%)': view_to_cart_dropout,
    'Cart to Purchase Dropout Rate (%)': cart_to_purchase_dropout,
    'Cart to Remove Dropout Rate (%)': cart_to_remove_dropout
}

dropout_rates
````

### ÏÉÅÌíàÎ≥Ñ Conversion Rate Î∂ÑÏÑù
---

````python
# ÏÉÅÌíàÎ≥Ñ Ï°∞Ìöå, Ïû•Î∞îÍµ¨Îãà Ï∂îÍ∞Ä, Íµ¨Îß§ Îç∞Ïù¥ÌÑ∞ Í≥ÑÏÇ∞
product_events = df.groupby(['product_id', 'event_type']).size().unstack().fillna(0)

# Íµ¨Îß§ Ï†ÑÌôòÎ•† Í≥ÑÏÇ∞: Íµ¨Îß§Ïàò / Ï°∞ÌöåÏàò
product_events['conversion_rate'] = product_events['purchase'] / product_events['view'] * 100

# ÏÉÅÌíàÎ≥Ñ Ï°∞ÌöåÏàòÍ∞Ä 10Ìöå Ïù¥ÏÉÅÏù∏ ÏÉÅÌíàÎì§Îßå ÌïÑÌÑ∞ÎßÅÌïòÏó¨ Ï†ÑÌôòÎ•† TOP 10 ÌôïÏù∏
top_conversion_products = product_events[product_events['view'] >= 10].sort_values(by=['view', 'conversion_rate'], ascending=[False, True]).head(10)

top_conversion_products[['view', 'cart', 'purchase', 'conversion_rate']]
````

### Ïùº(day)Î≥Ñ ÌñâÎèô Ìå®ÌÑ¥Î∂ÑÏÑù
---

````python
# event_timeÏóêÏÑú Ïùº Ï†ïÎ≥¥ Ï∂îÏ∂ú
def get_day(x): return x.day
df['day'] = df['event_time'].apply(get_day)
df.head()
````

````python
# ÏùºÎ≥Ñ Ïù¥Î≤§Ìä∏ Ïú†ÌòïÎ≥Ñ Î∞úÏÉù ÌöüÏàò Í≥ÑÏÇ∞
daily_events = df.groupby(['day', 'event_type']).size().unstack().fillna(0)

# ÏùºÎ≥Ñ Íµ¨Îß§ Ï†ÑÌôòÎ•† Í≥ÑÏÇ∞: Íµ¨Îß§Ïàò / Ï°∞ÌöåÏàò
daily_events['conversion_rate'] = daily_events['purchase'] / daily_events['view'] * 100

daily_events[['view', 'cart', 'purchase', 'conversion_rate']]
````

````python
# gradient ÏãúÍ∞ÅÌôî
styled_all_columns = daily_events.style.background_gradient(cmap='coolwarm', subset=['conversion_rate'])
styled_all_columns
````

### ÏÇ¨Ïö©Ïûê ÌñâÎèô Ìå®ÌÑ¥Î∂ÑÏÑù (Funnel Î∂ÑÏÑù)
---

````python
# ÌñâÎèôÌå®ÌÑ¥ Î∂ÑÏÑùÏùÑ ÏúÑÌïú Sorting
data_sorted = df.sort_values(['user_session', 'event_time'])
data_sorted.head()
````

````python
# DataÍ∞Ä ÎÑàÎ¨¥ ÌÅ¨ÎØÄÎ°ú Í∞ÄÏû• ÏùºÎ≥Ñ Íµ¨Îß§ Ï†ÑÌôòÎ•†(convention rate)Ïù¥ ÎÜíÏùÄ Ïùº(day)ÏùÑ Í∞ÄÏ†∏Ïò®Îã§
df_day = df[df['day'] == 12]
df_day['event_time'] = pd.to_datetime(df_day['event_time'])
df_day
````

````python
# ÏµúÏ¥à viewÎ•º ÏãúÍ∞ÑÍ∏∞Ï§ÄÏúºÎ°ú ÌïòÍ∏∞ÏúÑÌï¥ minÏúºÎ°úÏ†ïÎ†¨
grouped = df_day.groupby(['user_session', 'event_type'])["event_time"].min()
grouped
````

````python
# view, cart, purhase Îã®Í≥ÑÏàúÏÑúÎ•º ÏßÄÏ†ï
funnel_steps = pd.DataFrame({'steps':[1,2,3]}, index=['view','cart','purchase'])
funnel_steps
````

````python
# Funnel Îã®Í≥ÑÏôÄ merge
grouped = pd.DataFrame(grouped).merge(funnel_steps, left_on = 'event_type', right_index = True)
grouped
````

````python
# pivot tableÎ•º Ïù¥Ïö©ÌïòÏó¨ user sessionÎ≥Ñ Funnel Îã®Í≥Ñ ÏßÑÌñâÏó¨Î∂ÄÎ•º ÌôïÏù∏
funnel = grouped.reset_index().pivot(index='user_session',columns='steps',values='event_time')
funnel.columns = funnel_steps.index
funnel
````

````python
step_values = [funnel[column].notnull().sum() for column in funnel.columns]
step_values
````
